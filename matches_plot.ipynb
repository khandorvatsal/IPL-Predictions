### For the top7, only one team has won game whenever they've played. So we need to add another row to indicate the other team has won 0 matches.

```python
def match_plot(df):
    for a in df['Selector'].unique():
        b = df[df['Selector'] == a]
        if (len(b)==2):
            fig, ax = plt.subplots()
            ax = sns.barplot(x=b['level_1'], y=b['WINNER'])
        else:
            team2_name = a.replace(b['level_1'].iloc[0],'')
            fig, ax = plt.subplots()
            ax = sns.barplot(x=[b['level_1'].iloc[0], team2_name], y=[b["WINNER"].iloc[0], 0])
```

```python
match_plot(df)
```

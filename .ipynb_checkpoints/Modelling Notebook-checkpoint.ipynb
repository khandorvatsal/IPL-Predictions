```python
X = DataM.dropna().drop(['ID', 'CITY', 'CITY_Corrrected', 'TEAM1',
       'TEAM2', 'TOSS_DECISION', 'RESULT', 'WINNER','TOSS_WINNER'],axis=1)
X.describe()
```

# Train Test Splitting

```python
def data_split_by_year(df, split_year):
    index_drop=[]
    for year in range(split_year, 2019+1):
        index_drop = index_drop+df[df['SEASON']==year].index.values.tolist()
        
    X_train = df.drop(index_drop, axis=0).drop(['SEASON', 'WINNER_MODEL'],axis=1)
    y_train = df.drop(index_drop, axis=0)['WINNER_MODEL']      
    
    X_valid = df.loc[index_drop].drop(['SEASON', 'WINNER_MODEL'],axis=1)
    y_valid = df.loc[index_drop, 'WINNER_MODEL']
    
    return X_train,y_train,X_valid,y_valid
```

```python
%time X_train,y_train,X_valid,y_valid = data_split_by_year(X,2016)
```

# Modelling

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
def print_score(model,X_t,y_t,X_v,y_v):
    print(model.score(X_t,y_t),model.score(X_v,y_v))
```

```python
mod = RandomForestClassifier(n_estimators=200)
mod.fit(X_train, y_train)
```

```python
print_score(mod, X_train, y_train, X_valid, y_valid)
```

```python
mod.score(X_train, y_train)
```

```python
mod.score(X_valid, y_valid)
```

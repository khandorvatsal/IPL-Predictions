```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("final_data_with_individual_features_and_AHP.csv")
df
```

```python
df.drop(["Unnamed: 0", "Unnamed: 0.1", "ID", "TEAM1", "TEAM2", "CITY", "TOSS_DECISION", "WIN_BY_RUNS", "CITY_Corrrected", "TOSS_WINNER", "SEASON", "WINNER", "RESULT"], axis=1, inplace=True)
df
```

```python
df.dropna(axis=0, inplace=True)
df
```

```python
df.info()
```

```python
X = df.drop("WINNER_MODEL", axis=1)
y = df["WINNER_MODEL"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.2,
                                                    random_state=12)
```

```python
clf = LazyClassifier(verbose=0,ignore_warnings=True, custom_metric=None)
models,predictions = clf.fit(X_train, X_test, y_train, y_test)
models
```

#### Hence, decided to check SVM, DecisionTree, RidgeCV, AdaBoost, XGB, RFC

### SVC

```python
from sklearn.svm import SVC
clf1 = SVC()
clf1.fit(X_train, y_train)
```

```python
clf1.score(X_train, y_train)
```

```python
clf1.score(X_test, y_test)
```

### DecisionTree

```python
from sklearn.tree import DecisionTreeClassifier
clf2 = DecisionTreeClassifier()
clf2.fit(X_train, y_train)
```

```python
clf2.score(X_train, y_train)
```

```python
clf2.score(X_test, y_test)
```

### RFE

```python
from sklearn.feature_selection import RFE

score_train = list()
score_test = list()

for i in range(10, 50, 1):
    selector = RFE(DecisionTreeClassifier(), n_features_to_select=i, step=1)
    selector = selector.fit(X_train, y_train)
    
    score_train.append(selector.score(X_train, y_train))
    
    score_test.append(selector.score(X_test, y_test))
```

```python
score_train
```

```python
score_test
```

```python
max(score_test)
```

```python
import numpy as np
import pandas as pd

# Form a matrix of comparison.
df = pd.read_csv("ahp_1.csv")
df
```

```python
# We'll get rid of first column to make calculations easy.
df.drop("Unnamed: 0", axis=1, inplace=True)
df
```

```python
# Sum along a column
df.sum(axis=0)
```

```python
# Normalizing each column.
df["Overs"] = df["Overs"]/2.226189
df["Economy"] = df["Economy"]/3.816666
df["Wickets"] = df["Wickets"]/9.666666
df["Bowling Avg"] = df["Bowling Avg"]/17.2
df["Bowling SR"] = df["Bowling SR"]/17.25
df["4W Haul"] = df["4W Haul"]/29

df
```

```python
# Sum along each row
df["Sum"] = df.sum(axis=1)
df
```

```python
df["Criteria Weight"] = df["Sum"]/6
df
```

```python
# Checking Consistency
og = pd.read_csv("ahp_1.csv")
og.drop("Unnamed: 0", axis=1, inplace=True)
og
```

```python
# Multiply each column by respective criteria weight
og["Overs"] = og["Overs"]*0.387509
og["Economy"] = og["Economy"]*0.281308
og["Wickets"] = og["Wickets"]*0.158765
og["Bowling Avg"] = og["Bowling Avg"]*0.073609
og["Bowling SR"] = og["Bowling SR"]*0.067862
og["4W Haul"] = og["4W Haul"]*0.030947

og
```

```python
og.sum(axis=1)
```

```python
og["Weighted Sum"] = og.sum(axis=1)
og
```

```python
og["Criteria Weight"] = df["Criteria Weight"]
og
```

```python
og["Ratio of Sum and Criteria Weight"] = og["Weighted Sum"]/ og["Criteria Weight"]
og
```

```python
lambda_max = og["Ratio of Sum and Criteria Weight"].sum()/6
lambda_max
```

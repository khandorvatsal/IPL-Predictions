# Methods

```python
def replace_1_with_0(match_result):
    check_df = match_result.query('WINNER_MODEL == "0"').iloc [0:41,]
    check_df['temp'] = check_df['TEAM1']
    check_df['TEAM1'] = check_df['TEAM2']
    check_df['TEAM2'] = check_df['temp']
    check_df['WINNER'] = check_df['TEAM1']
    list_val = check_df.index
    match_result.loc[list_val] = check_df.loc[list_val]
    return match_result
```

```python
def delete_manually_columns():
    if 'TOSS_WINNER_MODEL' in match_result.columns: 
        match_result.drop(['TOSS_WINNER_MODEL'], axis=1, inplace = True)
    if 'WINNER_MODEL' in match_result.columns:
        match_result.drop(['WINNER_MODEL'], axis=1, inplace = True)
    if 'TOSS_DECISION_MODEL' in match_result.columns:
        match_result.drop(['TOSS_DECISION_MODEL'], axis=1, inplace = True)
    if 'CITY_MODEL' in match_result.columns:
        match_result.drop(['CITY_MODEL'], axis=1, inplace = True)
    if 'TEAM1_STRENGTH' in match_result.columns:
        match_result.drop(['TEAM1_STRENGTH'], axis=1, inplace = True)
    if 'TEAM2_STRENGTH' in match_result.columns:
        match_result.drop(['TEAM2_STRENGTH'], axis=1, inplace = True)
```

```python
def delete_manually_columns_strength():
    if 'TEAM1_STRENGTH' in win_rate.columns: 
        win_rate.drop(['TEAM1_STRENGTH'], axis=1, inplace = True)
    if 'TEAM2_STRENGTH' in win_rate.columns:
        win_rate.drop(['TEAM2_STRENGTH'], axis=1, inplace = True)
    if 'TEAM1_CUMU_STRENGTH' in win_rate.columns:
        win_rate.drop(['TEAM1_CUMU_STRENGTH'], axis=1, inplace = True)
    if 'TEAM2_CUMU_STRENGTH' in win_rate.columns:
        win_rate.drop(['TEAM2_CUMU_STRENGTH'], axis=1, inplace = True)
    
```

```python
def create_manually_columns(value):
    if value == "TOSS_WINNER":
        match_result.loc[match_result["TOSS_WINNER"]==match_result["TEAM2"],"TOSS_WINNER_MODEL"]= np.int64(0)
        match_result.loc[match_result["TOSS_WINNER"]==match_result["TEAM1"],"TOSS_WINNER_MODEL"]= np.int64(1)
        match_result["TOSS_WINNER_MODEL"] = match_result["TOSS_WINNER_MODEL"].astype(int)
    
    if value == "WINNER":
        match_result.loc[match_result["WINNER"]==match_result["TEAM2"],"WINNER_MODEL"]=np.int64(0)
        match_result.loc[match_result["WINNER"]==match_result["TEAM1"],"WINNER_MODEL"]=np.int64(1)
        match_result["WINNER_MODEL"] = match_result["WINNER_MODEL"].astype(int)
        
    if value == "TOSS_DECISION":
        match_result.loc[match_result["TOSS_DECISION"]== 'field',"TOSS_DECISION_MODEL"]=np.int64(0)
        match_result.loc[match_result["TOSS_DECISION"]!= 'field',"TOSS_DECISION_MODEL"]=np.int64(1)
        match_result["TOSS_DECISION_MODEL"] = match_result["TOSS_DECISION_MODEL"].astype(int)
        
    if value == "CITY_Corrrected":        
        match_result.loc[match_result["CITY_Corrrected"]== match_result["TEAM2"],"CITY_MODEL"]=np.int64(0)
        match_result.loc[match_result["CITY_Corrrected"]== match_result["TEAM1"],"CITY_MODEL"]=np.int64(1)
        match_result.loc[(match_result["CITY_Corrrected"]!= match_result["TEAM2"]) & (match_result["CITY_Corrrected"]!= match_result["TEAM1"]),"CITY_MODEL"]=np.int64(2)
        match_result["CITY_MODEL"] = match_result["CITY_MODEL"].astype(int)
   
```

```python
def missing_values_table(match_result):
        mis_val = match_result.isnull().sum()
        mis_val_percent = 100 * match_result.isnull().sum() / len(match_result)
        mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)
        mis_val_table_ren_columns = mis_val_table.rename(
        columns = {0 : 'Missing Values', 1 : '% of Total Values'})
        mis_val_table_ren_columns = mis_val_table_ren_columns[
            mis_val_table_ren_columns.iloc[:,1] != 0].sort_values(
        '% of Total Values', ascending=False).round(1)
        print ("Your selected dataframe has " + str(match_result.shape[1]) + " columns.\n"      
            "There are " + str(mis_val_table_ren_columns.shape[0]) +
              " columns that have missing values.")
        return mis_val_table_ren_columns

```

# Pipeline

```python
strength_features = ['Team Strength','Cumulative Team Strength']
strength_transformer = Pipeline(steps=[
    # We can add anything else 
    # We can manipulate more scalers here
    ('scaler', StandardScaler())])
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', strength_transformer, strength_features)
    ],remainder="passthrough")
scaled_strength = preprocessor.fit_transform(strength)
strength_scaled = pd.DataFrame(scaled_strength,columns = ['Team Strength','Cumulative Team Strength','Team','Year'])
strength_scaled.head()
```

```python
strength_scaled['Next_Year'] = strength_scaled['Year'] + 1
```

```python
TEAM1_STRENGTH = pd.Series([]) 
TEAM2_STRENGTH = pd.Series([])
TEAM1_CUMU_STRENGTH = pd.Series([]) 
TEAM2_CUMU_STRENGTH = pd.Series([])
for(i,match) in win_rate.iterrows():
    for(j,stre) in strength_scaled.iterrows():
        if match['TEAM1'] == stre['Team']  and (int(match['SEASON'])) == (int(stre['Year']) + 1):
            TEAM1_STRENGTH[i] = stre['Team Strength']
            TEAM1_CUMU_STRENGTH[i] = stre['Cumulative Team Strength']
        if match['TEAM2'] == stre['Team']  and int(match['SEASON']) == (int(stre['Year']) + 1):
            TEAM2_STRENGTH[i] = stre['Team Strength']
            TEAM2_CUMU_STRENGTH[i] =stre['Cumulative Team Strength']
```

```python
#correlation heatmap of dataset
def correlation_heatmap(df):
    _ , ax = plt.subplots(figsize =(14, 12))
    colormap = sns.diverging_palette(220, 10, as_cmap = True)
    
    _ = sns.heatmap(
        df.corr(), 
        cmap = colormap,
        square=True, 
        cbar_kws={'shrink':.9 }, 
        ax=ax,
        annot=True, 
        linewidths=0.1,vmax=1.0, linecolor='white',
        annot_kws={'fontsize':12 }
    )
    
    plt.title('Pearson Correlation of Features', y=1.05, size=15)

```

# fixing team names

```python
data.replace({'TEAM1':{'RPSs':'RPS','PW':'RPS','DC':'SRH'}},inplace=True)
data.replace({'TEAM2':{'RPSs':'RPS','PW':'RPS','DC':'SRH'}},inplace=True)
data.replace({'WINNER':{'RPSs':'RPS','PW':'RPS','DC':'SRH'}},inplace=True)
data.replace({'TOSS_WINNER':{'RPSs':'RPS','PW':'RPS','DC':'SRH'}},inplace=True)
```

# creating new rows

```python
null_values = []
for i in range(len(data)):
    null_values.append(0)
data['WIN_RATE_TEAM1'] = null_values
data['WIN_RATE_TEAM2'] = null_values
```

# code for calculating win rate

```python


def update_win_rate(dataframe):
    match_wins_team_wise = {'KKR':0,'CSK':0,'RCB':0,'KXIP':0,'DD':0,'SRH':0,'RPS':0,'GL':0,'KTK':0,'RR':0,'MI':0}
    for j in range(len(dataframe)):
        if(dataframe.loc[j,'SEASON'] == 2008):
            match_wins_team_wise[dataframe.loc[j,'WINNER']] +=1
    years = [2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019]
    for i in range(len(years)-1): 
        for k in range(len(dataframe)):
            if(dataframe.loc[k,'SEASON'] == years[i+1]):
                dataframe.loc[k,'WIN_RATE_TEAM1'] = match_wins_team_wise[dataframe.loc[k,'TEAM1']]/ipl_match_numbers_by_year[years[i]]
                dataframe.loc[k,'WIN_RATE_TEAM2'] = match_wins_team_wise[dataframe.loc[k,'TEAM2']]/ipl_match_numbers_by_year[years[i]]
        for j in range(len(dataframe)):
            if(dataframe.loc[j,'SEASON'] == years[i+1]):
                match_wins_team_wise[dataframe.loc[j,'WINNER']] +=1
    return dataframe 
```

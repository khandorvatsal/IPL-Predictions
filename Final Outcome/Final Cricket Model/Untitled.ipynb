```python
def read_player_data():
    path = r'Auction_data'
    all_files = glob.glob(path + "/*.xlsx")
    
    li = []

    for filename in all_files:
        df = pd.read_excel(filename, index_col=None, header=0)
        #df = pd.read_excel('Auction_data/auction_data_2008.xlsx')
        df = df.iloc[:,[2,6,7]]
        df.columns = ['Name', ' Team' , 'Dollar']
        temp = re.findall(r'\d+', filename) 
        df['Year'] = temp[0]
        li.append(df)
        break

    all_years_player_performance = pd.concat(li, axis=0, ignore_index=True)
    return all_years_player_performance
```

## Now that we've done our both bat and bowl, we'll have to merge this manually from the csv's exported and then do groupby and then add points column manually again.

```python
# We've created 2019_overall from the both csv exported. Let's work on this !!
```

```python
final_2019 = pd.read_csv("2019_overall.csv")
final_2019
```

```python
final_2019.info()
```

```python
# Since we have 162 in batsman's list and 162 in bowler's list, it should be twice occurrence of a player.
final_2019["Player"].value_counts()
```

### Yeah we we're correct !!!!

```python
final_2019_merged = final_2019.groupby("Player").sum() 
```

```python
final_2019_merged
```

```python
final_2019_merged = final_2019_merged.reset_index()
final_2019_merged
```

### Now we'll export this and add player points manually.

```python
# We'll comment after first run otherwise it would edit the file on which we've done manual editting.
# final_2019_merged.to_csv("final_2019_merged.csv", index=False)
```

```python
final_2019_merged = pd.read_csv("final_2019_merged.csv")
final_2019_merged
```

```python
final_2019_merged.info()
```

```python
X = final_2019_merged.drop(["Player", "Points"], axis=1)
y = final_2019_merged["Points"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.2,
                                                    random_state=12)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
model.score(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
y_test = np.array(y_test)
y_preds = np.array(model.predict(X_test))
```

```python
np.set_printoptions(precision=2)
print(np.concatenate((y_preds.reshape(len(y_preds),1), y_test.reshape(len(y_test),1)),1))
```
